```python

import face_recognition
import cv2
import numpy as np
import csv 
from datetime import datetime
import pandas as pd

video_capture = cv2.VideoCapture(0)

# Load known faces

alakh_image = face_recognition.load_image_file("D:\Attendence system - Minor Project/students/Alakh.jpeg")
sunny_deol_image = face_recognition.load_image_file("D:\Attendence system - Minor Project/students/Sunny-Deol.jpeg")
akarsh_image = face_recognition.load_image_file(r"D:\Attendence system - Minor Project\students/Akarsh.jpg")

#D:\Attendence system - Major Project\students


alakh_encoding = face_recognition.face_encodings(alakh_image)[0]
sunny_deol_encoding = face_recognition.face_encodings(sunny_deol_image)[0]

akarsh_encoding= face_recognition.face_encodings(akarsh_image)[0]





known_face_encodings = [alakh_encoding , sunny_deol_encoding , akarsh_encoding]
known_face_names = ["Alakh" , 'Sunny Deol'," Akarsh "]

# List of expected students

students = known_face_names.copy()

face_locations = []
face_encodings = []

# Get the current date and time

now = datetime.now()
current_date = now.strftime("%Y-%m-%d")

f = open(f"{current_date}.csv" , "w+" , newline="")
lnwriter = csv.writer(f)

while True:
    _, frame = video_capture.read()
    small_frame = cv2.resize(frame ,(0,0) , fx=0.25 , fy=0.25)
    rgb_small_frame = cv2.cvtColor(small_frame , cv2.COLOR_BGR2RGB) 
    
    # Recognise faces
    face_locations = face_recognition.face_locations(rgb_small_frame)
    face_encodings = face_recognition.face_encodings(rgb_small_frame , face_locations)
    
    for face_encoding in face_encodings:
        matches = face_recognition.compare_faces(known_face_encodings,face_encoding)
        face_distance = face_recognition.face_distance(known_face_encodings,face_encoding)
        best_match_index = np.argmin(face_distance)
        
        if(matches[best_match_index]):
            name = known_face_names[best_match_index]
            
            # Add the text if a person is present
            if name in known_face_names:
                font = cv2.FONT_HERSHEY_SIMPLEX
                bottomLeftCornerOfText = (10,100)
                fontScale = 1.5
                fontColor = (255,0,0)
                thickness = 3
                lineType = 2
                cv2.putText(frame, name + " Present", bottomLeftCornerOfText, font, fontScale, fontColor, thickness, lineType)
                
                if name in students:
                    students.remove(name)
                    current_time = now.strftime("%H:%M:%S")
                    lnwriter.writerow([name,current_time])
    
    cv2.imshow("Camera" , frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break
    
video_capture.release()
cv2.destroyAllWindows()
f.close()
```
